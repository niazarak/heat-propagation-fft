```python
import numpy as np
import matplotlib.pyplot as plt
import pathlib
import csv
import itertools
```

```python
d = pathlib.Path('out/')
files = sorted(list(d.iterdir()), key=lambda f: float(f.name[:-4]))
```

```python
datas = []
for data_f in files:
    data = []
    f = data_f.open()
    reader = csv.reader(f, delimiter=' ')
    for r in reader:
        row = ([float(n) for n in r if n])
        if row:
            data += row
    f.close()
    datas.append(np.array(data))
N = int(len(datas[0]) ** 0.5)
```

```python
def print_heat(t):
    x = list(itertools.product(range(N), range(N)))
    y = datas[t]
    plt.scatter([_x[0] for _x in x], [_x[1] for _x in x], c = y)
    print(datas[t])
```

```python
print_heat(0)
```

```python
print_heat(5)
```

```python
print_heat(49)
```
